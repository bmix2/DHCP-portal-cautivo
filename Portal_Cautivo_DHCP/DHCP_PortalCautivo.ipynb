# **DESARROLLO:**

**1)¿Qué es DHCP?**

El Protocolo de configuración dinámica de host (DHCP) es un protocolo de red que utiliza una arquitectura cliente/servidor que proporciona automáticamente un host de Protocolo de Internet (IP) con su dirección IP y otra información de configuración relacionada, como la máscara de subred y la puerta de enlace predeterminada.
Este servidor posee una lista de direcciones IP dinámicas y las va asignando a los clientes conforme estas van quedando libres, sabiendo en todo momento quién ha estado en posesión de esa IP, cuánto tiempo la ha tenido y a quién se la ha asignado después. Así los clientes de una red IP pueden conseguir sus parámetros de configuración automáticamente.


**2) ¿Qué es Portal Cautivo?**

Un portal cautivo es un programa o máquina de una red informática que vigila el tráfico HTTP y fuerza a los usuarios a pasar por una página especial si quieren navegar por Internet de forma normal. El portal cautivo es una página web, también conocida como una página de login, en la cual el usuario ve antes de acceder la red Wi-Fi pública donde los usuarios pueden autenticarse utilizando sus propias credenciales.


**3) ¿Herramientas de implementación**

Utilizaremos un entorno virtual para simular una implementacion de estos servicios, es decir montaremos un laboratorio de pruebas, por medio de maquinas virtuales en donde una de ellas llevara el servidor DHCP y de portal cautivo, por otro lado las otras maquinas consumiran los servicios mencionados anteriormente.

Para este laboratorio nos apoyaremos de una herramienta digital llamada PfSense el cual nos brinda configuraciones para la implementación del DHCP, DNS, trafico HTTP y Portal Cautivo, siendo que nos centraremos en los dos protocolos requeridos (DHCP - Portal Cautivo), dicha herramienta la encontraremos en el siguiente link: https://www.pfsense.org/download/

Aunado a esto, usaremos VirtualBox como software para la instalación del ISO PfSense, el cliente (el cual puede ser de cualquier tipo, Windows, Linux, etc.), y en el mismo, las pruebas y verificación del servicio corriendo. Podemos descargar VirtualBox en el siguiente link: https://www.virtualbox.org/wiki/Downloads
